```python
a = ("Age")
b = ("Pin code")
```

```python
fig, ax = plt.subplots()

ax.plot(df[a], df[b])
ax.set(xlabel = 'occupation', ylabel = 'Gender', title = 'occupation vs Gender')

ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(df[a], df[b])
ax.set(xlabel = 'occupation', ylabel = 'Gender', title = 'occupation vs Gender')

ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

gender_counts = df['Gender'].value_counts()

ax.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
ax.set(title = 'Distribution of Gender')


plt.show()
```
